```python
# Training dataset
labels = df_train.drop(columns=['y'])
target = df_train['y']

# Test dataset
labels_test = df_test.drop(columns=['y'])
target_test = df_test['y']

# Split training and validation dataset
target_train, target_temp, labels_train, labels_temp = train_test_split(
    target, labels, test_size=0.20, random_state=42)

# Preprocessing
X_train = labels_train
X_val = labels_temp
X_test = labels_test

# Separate the target
y_train = pd.Series(target_train)
y_val = pd.Series(target_temp)
y_test = pd.Series(target_test)
```

```python
# Define the base model
model = CatBoostClassifier(verbose=0)

# Hyperparameter grid
param_grid = {
    'iterations': [200, 300, 400],
    'depth': [5, 6, 7],
    'learning_rate': [0.01, 0.1, 0.3],
    'l2_leaf_reg': [1, 3, 5],
}

# Set up scorer and GridSearchCV
scorer = make_scorer(recall_score, pos_label=0)
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring=scorer, n_jobs=-1)

# Train with GridSearch
grid_search.fit(X_train, y_train)

# Display results
print(f"Best hyperparameters: {grid_search.best_params_}")
```

```python
# Start training time measurement
start_time = time.time()

# Use the best model and make predictions
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
y_pred_proba = best_model.predict_proba(X_test)[:, 1]

# Calculate training time
training_time = time.time() - start_time

# Define threshold and make adjusted predictions
threshold = 0.67
y_pred_threshold = (y_pred_proba >= threshold).astype(int)

# Calculate metrics
report = classification_report(y_test, y_pred_threshold, output_dict=True)
precision_neg, recall_neg, f1_neg, _ = precision_recall_fscore_support(y_test, y_pred_threshold, pos_label=0, average=None)
precision_pos, recall_pos, f1_pos, _ = precision_recall_fscore_support(y_test, y_pred_threshold, pos_label=1, average=None)

# Extract values by class
precision_neg, precision_pos = precision_neg[0], precision_pos[1]
recall_neg, recall_pos = recall_neg[0], recall_pos[1]
f1_neg, f1_pos = f1_neg[0], f1_pos[1]

# Calculate overall metrics
roc_auc = roc_auc_score(y_test, y_pred_proba)
logloss = log_loss(y_test, y_pred_proba)
conf_matrix = confusion_matrix(y_test, y_pred_threshold)
conf_matrix_norm = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis]
accuracy = accuracy_score(y_test, y_pred_threshold)

# Display results
print("Metrics by Class:")
print(f"Class 0 - Precision: {precision_neg:.2f}, Recall: {recall_neg:.2f}, F1-Score: {f1_neg:.2f}")
print(f"Class 1 - Precision: {precision_pos:.2f}, Recall: {recall_pos:.2f}, F1-Score: {f1_pos:.2f}")
print("\nOverall Metrics:")
print(f"Accuracy: {accuracy:.2f}")
print(f"ROC AUC: {roc_auc:.2f}")
print(f"Log-Loss: {logloss:.2f}")
print("\nNormalized Confusion Matrix:")
print("\nTraining Time:")
print(f"Time elapsed: {training_time:.2f} seconds")
print(conf_matrix_norm)
```

```python
# Financial parameters
average_loan_amount = 10000
return_rate = 0.20  # 20% return on paid loans
loss_rate = 0.70    # 70% loss on defaulted loans

# Calculate financial scenarios
tp = sum((y_test == 1) & (y_pred_threshold == 1))
tn = sum((y_test == 0) & (y_pred_threshold == 0))
fp = sum((y_test == 0) & (y_pred_threshold == 1))
fn = sum((y_test == 1) & (y_pred_threshold == 0))

# Calculate financial impacts
tp_gain = tp * average_loan_amount * return_rate
fp_loss = fp * average_loan_amount * loss_rate
fn_opportunity_loss = fn * average_loan_amount * return_rate

total_gain = tp_gain
total_loss = abs(fp_loss + fn_opportunity_loss)
net_result = total_gain - total_loss

# Calculate ROI
total_predicted_positives = sum(y_pred_threshold == 1)
roi = (net_result / (total_predicted_positives * average_loan_amount)) * 100

# Print results
print("\nFinancial Analysis:")
print(f"Parameters:")
print(f"- Average loan amount: ${average_loan_amount:,.2f}")
print(f"- Return rate: {return_rate*100}%")
print(f"- Loss rate: {loss_rate*100}%")

print("\nResults by Scenario:")
print(f"1. True Positives (TP):")
print(f"   - Count: {tp:,}")
print(f"   - Total Gain: ${tp_gain:,.2f}")
print(f"   - Average Gain per Operation: ${tp_gain/max(tp, 1):,.2f}")

print(f"\n2. True Negatives (TN):")
print(f"   - Count: {tn:,}")
print(f"   - Avoided Losses: ${tn * average_loan_amount * loss_rate:,.2f}")

print(f"\n3. False Positives (FP):")
print(f"   - Count: {fp:,}")
print(f"   - Total Loss: ${fp_loss:,.2f}")
print(f"   - Average Loss per Operation: ${fp_loss/max(fp, 1):,.2f}")

print(f"\n4. False Negatives (FN):")
print(f"   - Count: {fn:,}")
print(f"   - Missed Opportunities: ${fn_opportunity_loss:,.2f}")
print(f"   - Average Loss per Operation: ${fn_opportunity_loss/max(fn, 1):,.2f}")

print("\nOverall Financial Results:")
print(f"- Total Gain: ${total_gain:,.2f}")
print(f"- Total Loss: ${total_loss:,.2f}")
print(f"- Net Result: ${net_result:,.2f}")
print(f"- ROI: {roi:.2f}%")

# Visualizations
plt.figure(figsize=(15, 5))

# Financial results
plt.subplot(121)
scenario_results = [tp_gain, -fp_loss, -fn_opportunity_loss]
plt.bar(['Gains (TP)', 'Losses (FP)', 'Missed Opp. (FN)'],
       scenario_results,
       color=['green', 'red', 'orange'])
plt.title('Financial Results by Scenario')
plt.ylabel('Value ($)')
plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)

# Prediction distribution
plt.subplot(122)
plt.pie([tp, tn, fp, fn], 
       labels=['TP', 'TN', 'FP', 'FN'],
       colors=['green', 'blue', 'red', 'orange'],
       autopct='%1.1f%%')
plt.title('Prediction Distribution')

plt.tight_layout()
plt.show()
```

```python
def load_data(train_path, test_path, sample_size=None, random_state=42):
    """
    Loads training and test data with optional sampling
    
    Args:
        train_path: Path to training data CSV
        test_path: Path to test data CSV 
        sample_size: Optional size to sample from training data
        random_state: Random seed for reproducibility
    
    Returns:
        Tuple of (train_df, test_df) as pandas DataFrames
    """
    print("[INFO] Loading data...")
    df_train = pd.read_csv(train_path)
    df_test = pd.read_csv(test_path)

    if sample_size and sample_size < len(df_train):
        df_train = df_train.sample(sample_size, random_state=random_state)
        print(f"[INFO] Using sample of {sample_size} records for initial analysis")

    print("[INFO] Data loaded successfully.")
    return df_train, df_test

def complete_data_exploration(df, dataset_name, sample_size=10000):
    """
    Performs comprehensive exploratory data analysis
    
    Args:
        df: Input DataFrame
        dataset_name: Name of dataset for reporting
        sample_size: Max rows to analyze
        
    Generates visualizations and prints statistics about the data
    """
    df = df.sample(min(sample_size, len(df)), random_state=42)
    print(f"\n[INFO] Complete exploration of {dataset_name} dataset:")
    
    # Basic info
    print("\n1. Basic Information:")
    print(f"Number of rows and columns: {df.shape}")
    print(df.info())
    
    # Descriptive stats
    print("\n2. Descriptive Statistics:")
    print(df.describe().T)
    
    # Missing values analysis
    print("\n3. Missing Values Analysis:")
    missing_values = df.isnull().sum()
    missing_percentages = 100 * missing_values / len(df)
    missing_table = pd.concat([missing_values, missing_percentages], axis=1, keys=['Total', 'Percentage'])
    print(missing_table[missing_table['Total'] > 0].sort_values('Total', ascending=False))
    
    # Correlation matrix
    print("\n4. Correlation Matrix:")
    plt.figure(figsize=(12, 10))
    corr_matrix = df.corr()
    sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
    plt.title('Variable Correlation Matrix')
    plt.show()
    
    # Top correlations
    print("\nTop 10 strongest correlations:")
    corr_pairs = corr_matrix.unstack()
    corr_pairs = corr_pairs[corr_pairs != 1.0].sort_values(ascending=False)
    print(corr_pairs.head(10))
    
    # Check anonymization
    identifiable_cols = [col for col in df.columns if 'id' in col.lower() or 'name' in col.lower() or 'ssn' in col.lower()]
    if identifiable_cols:
        print(f"\n[WARNING] Dataset contains potentially identifiable data: {identifiable_cols}")
    else:
        print("\n[INFO] Data appears to be anonymized.")
    print("\n[INFO] Data exploration completed.")

def analyze_target_distribution(df):
    """
    Analyzes and visualizes target variable distribution
    """
    print("[INFO] Analyzing target distribution...")
    plt.figure(figsize=(8, 6))
    sns.countplot(x='y', data=df)
    plt.title('Target Variable (y) Distribution')
    plt.show()
    print(f"Target distribution:\n{df['y'].value_counts(normalize=True)}")

def analyze_correlations(df):
    """
    Analyzes feature correlations and identifies highly correlated pairs
    
    Returns:
        Set of variables recommended for removal
    """
    print("[INFO] Calculating correlation matrix...")
    corr_matrix = df.corr()
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
    plt.title('Variable Correlation Matrix')
    plt.show()
    
    high_corr = corr_matrix.abs().unstack().sort_values(ascending=False)
    high_corr = high_corr[(high_corr > 0.8) & (high_corr < 1.0)]
    print("\n[INFO] Highly correlated variables (correlation > 0.8):")
    print(high_corr)
    
    variables_to_remove = set()
    for (var1, var2), corr_value in high_corr.items():
        if var1 != var2:
            variables_to_remove.add(var1)
            print(f"[SUGGESTION] Consider removing one of correlated variables: {var1} or {var2}")
    return variables_to_remove

def handle_missing_values(df_train, df_test):
    """
    Handles missing values consistently between train and test sets
    using median imputation
    
    Returns:
        Tuple of imputed (train_df, test_df)
    """
    print("[INFO] Handling missing values...")
    print("\n[INFO] Missing values per column (train):")
    print(df_train.isnull().sum())
    
    common_cols = df_train.columns.intersection(df_test.columns)
    df_train = df_train[common_cols]
    df_test = df_test[common_cols]

    imputer = SimpleImputer(strategy='median')
    df_train_imputed = pd.DataFrame(
        imputer.fit_transform(df_train),
        columns=df_train.columns,
        index=df_train.index
    )
    df_test_imputed = pd.DataFrame(
        imputer.transform(df_test),
        columns=df_test.columns,
        index=df_test.index
    )

    print("[INFO] Missing values handled successfully.")
    return df_train_imputed, df_test_imputed

def handle_outliers(df):
    """
    Handles outliers using IQR method
    """
    print("[INFO] Handling outliers...")
    num_columns = df.columns.drop('y', errors='ignore')

    for col in num_columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df[col] = df[col].clip(lower_bound, upper_bound)
    print("[INFO] Outliers handled successfully.")
    return df

def analyze_missing_values_detailed(df_train, df_test):
    """
    Performs detailed analysis of missing values and their impact on target
    
    Returns:
        Dict containing impact analysis and recommendations
    """
    results = {}
    na_counts = df_train.isnull().sum()
    na_percentages = (na_counts / len(df_train)) * 100

    impact_analysis = {}
    for column in df_train.columns:
        if column != 'y' and df_train[column].isnull().sum() > 0:
            is_missing = df_train[column].isnull()
            default_rate_present = df_train[~is_missing]['y'].mean()
            default_rate_missing = df_train[is_missing]['y'].mean()
            contingency = pd.crosstab(is_missing, df_train['y'])
            chi2, p_value = stats.chi2_contingency(contingency)[:2]

            impact_analysis[column] = {
                'na_percentage': na_percentages[column],
                'default_rate_present': default_rate_present,
                'default_rate_missing': default_rate_missing,
                'rate_difference': abs(default_rate_present - default_rate_missing),
                'p_value': p_value
            }

    features_to_keep = []
    features_to_drop = []
    recommendations = []

    for column, stats_dict in impact_analysis.items():
        if stats_dict['na_percentage'] > 50:
            if stats_dict['p_value'] < 0.05 and stats_dict['rate_difference'] > 0.1:
                features_to_keep.append(column)
                recommendations.append(f"Keep {column} despite high NA - Significant target impact")
            else:
                features_to_drop.append(column)
                recommendations.append(f"Remove {column} - High NA without significant impact")

    print("\n=== Missing Values Analysis ===")
    print(f"\nFeatures with >50% NA: {len([col for col, val in na_percentages.items() if val > 50])}")
    print(f"Features with significant impact: {len(features_to_keep)}")

    print("\nTop 5 features with highest target impact:")
    impact_df = pd.DataFrame(impact_analysis).T
    print(impact_df.sort_values('rate_difference', ascending=False).head())

    return {
        'impact_analysis': impact_analysis,
        'features_to_keep': features_to_keep,
        'features_to_drop': features_to_drop,
        'recommendations': recommendations
    }
```

```python
def plot_learning_curve(estimator, X, y, title):
    """
    Plot learning curves to visualize model's performance with varying training set sizes.
    
    Args:
        estimator: The classifier/model to analyze
        X: Features matrix
        y: Target vector
        title: Title for the plot
    """
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=5, n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 5))
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.figure(figsize=(10, 6))
    plt.title(title)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    plt.grid()
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
    plt.legend(loc="best")
    plt.show()

def generate_report(final_model, X_test, y_test, financial_metrics, training_time):
    """
    Generate comprehensive report including model performance and financial metrics.
    
    Args:
        final_model: Trained model
        X_test: Test features
        y_test: Test labels
        financial_metrics: Dict containing financial performance metrics
        training_time: Model training duration in seconds
    """
    print("[INFO] Generating final report...")
    with open("adaboost_complete_report.md", "w") as f:
        f.write("# Credit Risk Analysis Report using AdaBoost\n\n")

        # Model Configuration
        f.write("## 1. Model Configuration\n")
        f.write(f"- Number of Estimators: {final_model.n_estimators}\n")
        f.write(f"- Learning Rate: {final_model.learning_rate}\n")
        f.write(f"- Algorithm: {final_model.algorithm}\n")
        f.write(f"- Training Time: {training_time:.2f} seconds\n\n")

        # Classification Metrics
        f.write("## 2. Classification Metrics\n")
        y_pred = final_model.predict(X_test)

        # Class 0 (Default) Metrics
        f.write("### Class 0 (Default)\n")
        f.write(f"- Accuracy: {accuracy_score(y_test == 0, y_pred == 0):.4f}\n")
        f.write(f"- Precision: {precision_score(y_test, y_pred, pos_label=0):.4f}\n")
        f.write(f"- Recall: {recall_score(y_test, y_pred, pos_label=0):.4f}\n")
        f.write(f"- F1-Score: {f1_score(y_test, y_pred, pos_label=0):.4f}\n\n")

        # Class 1 (Non-Default) Metrics
        f.write("### Class 1 (Non-Default)\n")
        f.write(f"- Accuracy: {accuracy_score(y_test == 1, y_pred == 1):.4f}\n")
        f.write(f"- Precision: {precision_score(y_test, y_pred, pos_label=1):.4f}\n")
        f.write(f"- Recall: {recall_score(y_test, y_pred, pos_label=1):.4f}\n")
        f.write(f"- F1-Score: {f1_score(y_test, y_pred, pos_label=1):.4f}\n\n")

        # Financial Metrics
        f.write("## 3. Financial Analysis\n")
        f.write(f"- Total Gain: $ {financial_metrics['total_gain']:,.2f}\n")
        f.write(f"- Total Loss: $ {financial_metrics['total_loss']:,.2f}\n")
        f.write(f"- Net Result: $ {financial_metrics['net_result']:,.2f}\n")
        f.write(f"- ROI: {financial_metrics['roi']:.2f}%\n\n")

        # Scenario Analysis
        f.write("## 4. Scenario Analysis\n")

        # Class 0 Scenarios
        f.write("### Class 0 (Default)\n")
        tp_0 = sum((y_test == 0) & (y_pred == 0))
        tn_0 = sum((y_test == 1) & (y_pred == 1))
        fp_0 = sum((y_test == 1) & (y_pred == 0))
        fn_0 = sum((y_test == 0) & (y_pred == 1))

        f.write(f"- True Negatives (Predicted 0, Actual 0): {tp_0:,}\n")
        f.write(f"- False Negatives (Predicted 1, Actual 0): {fn_0:,}\n")
        f.write(f"- Avoided Losses: $ {tp_0 * 10000 * 0.7:,.2f}\n\n")

        # Class 1 Scenarios
        f.write("### Class 1 (Non-Default)\n")
        f.write(f"- True Positives (TP): {financial_metrics['tp']:,}\n")
        f.write(f"- Total Gain: $ {financial_metrics['total_gain']:,.2f}\n")

        f.write("\n### True Negatives (TN)\n")
        f.write(f"- Count: {financial_metrics['tn']:,}\n")
        f.write(f"- Avoided Losses: $ {financial_metrics['tn'] * 10000 * 0.7:,.2f}\n")

        f.write("\n### False Positives (FP)\n")
        f.write(f"- Count: {financial_metrics['fp']:,}\n")
        f.write(f"- Total Loss: $ {financial_metrics['total_loss']:,.2f}\n")

        f.write("\n### False Negatives (FN)\n")
        f.write(f"- Count: {financial_metrics['fn']:,}\n")
        f.write(f"- Missed Opportunities: $ {financial_metrics['fn_opportunity_loss']:,.2f}\n")

        # Class Distribution
        f.write("\n## 5. Class Distribution\n")
        total_samples = len(y_test)
        class_0_count = sum(y_test == 0)
        class_1_count = sum(y_test == 1)

        f.write("### Actual Distribution\n")
        f.write(f"- Class 0 (Default): {class_0_count:,} ({(class_0_count/total_samples)*100:.2f}%)\n")
        f.write(f"- Class 1 (Non-Default): {class_1_count:,} ({(class_1_count/total_samples)*100:.2f}%)\n\n")

        pred_class_0_count = sum(y_pred == 0)
        pred_class_1_count = sum(y_pred == 1)

        f.write("### Predicted Distribution\n")
        f.write(f"- Class 0 (Default): {pred_class_0_count:,} ({(pred_class_0_count/total_samples)*100:.2f}%)\n")
        f.write(f"- Class 1 (Non-Default): {pred_class_1_count:,} ({(pred_class_1_count/total_samples)*100:.2f}%)\n\n")

        # Conclusions
        f.write("\n## 6. Conclusions and Recommendations\n")
        f.write("- Overall model performance analysis\n")
        f.write("- Financial impact of decisions\n")
        f.write("- Suggestions for future improvements\n")

    print("[INFO] Report generated successfully: adaboost_complete_report.md")

def calculate_financial_metrics(y_true, y_pred, average_loan_amount=10000):
    """Calculate financial metrics to assess monetary impact of predictions"""
    results = pd.DataFrame({
        'actual': y_true,
        'prediction': y_pred
    })

    return_rate = 0.20
    loss_rate = 0.70

    def calculate_result(row):
        if row['actual'] == 1 and row['prediction'] == 1:    # True Positive
            return average_loan_amount * return_rate
        elif row['actual'] == 0 and row['prediction'] == 0:  # True Negative
            return 0
        elif row['actual'] == 0 and row['prediction'] == 1:  # False Positive
            return -average_loan_amount * loss_rate
        else:  # False Negative
            return -average_loan_amount * return_rate

    results['financial_result'] = results.apply(calculate_result, axis=1)

    metrics = {
        'total_loans': len(y_true),
        'total_positive_predictions': sum(y_pred == 1),
        'total_negative_predictions': sum(y_pred == 0),

        'tp': sum((y_true == 1) & (y_pred == 1)),
        'tn': sum((y_true == 0) & (y_pred == 0)),
        'fp': sum((y_true == 0) & (y_pred == 1)),
        'fn': sum((y_true == 1) & (y_pred == 0)),

        'total_gain': results[results['financial_result'] > 0]['financial_result'].sum(),
        'total_loss': abs(results[results['financial_result'] < 0]['financial_result'].sum()),
        'net_result': results['financial_result'].sum(),

        'tp_gain': sum((y_true == 1) & (y_pred == 1)) * average_loan_amount * return_rate,
        'fp_loss': sum((y_true == 0) & (y_pred == 1)) * average_loan_amount * loss_rate,
        'fn_opportunity_loss': sum((y_true == 1) & (y_pred == 0)) * average_loan_amount * return_rate
    }

    metrics['roi'] = (metrics['net_result'] /
                     (metrics['total_positive_predictions'] * average_loan_amount) * 100)

    print("\n=== Detailed Financial Analysis ===")
    print(f"\nParameters used:")
    print(f"- Average loan amount: $ {average_loan_amount:,.2f}")
    print(f"- Return rate (paid loans): {return_rate*100}%")
    print(f"- Loss rate (defaulted loans): {loss_rate*100}%")

    print("\nResults by Scenario:")
    print(f"1. True Positives (TP):")
    print(f"   - Count: {metrics['tp']:,}")
    print(f"   - Total Gain: $ {metrics['tp_gain']:,.2f}")
    print(f"   - Average Gain per Operation: $ {metrics['tp_gain']/max(metrics['tp'], 1):,.2f}")

    print(f"\n2. True Negatives (TN):")
    print(f"   - Count: {metrics['tn']:,}")
    print(f"   - Avoided Losses: $ {metrics['tn'] * average_loan_amount * loss_rate:,.2f}")

    print(f"\n3. False Positives (FP):")
    print(f"   - Count: {metrics['fp']:,}")
    print(f"   - Total Loss: $ {metrics['fp_loss']:,.2f}")
    print(f"   - Average Loss per Operation: $ {metrics['fp_loss']/max(metrics['fp'], 1):,.2f}")

    print(f"\n4. False Negatives (FN):")
    print(f"   - Count: {metrics['fn']:,}")
    print(f"   - Missed Opportunities: $ {metrics['fn_opportunity_loss']:,.2f}")
    print(f"   - Average Loss per Operation: $ {metrics['fn_opportunity_loss']/max(metrics['fn'], 1):,.2f}")

    print("\nGlobal Results:")
    print(f"- Total Gain: $ {metrics['total_gain']:,.2f}")
    print(f"- Total Loss: $ {metrics['total_loss']:,.2f}")
    print(f"- Net Result: $ {metrics['net_result']:,.2f}")
    print(f"- ROI: {metrics['roi']:.2f}%")

    visualize_financial_metrics(metrics)
    
    return metrics

def visualize_financial_metrics(metrics):
    """Visualize financial metrics with plots"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    scenario_results = [
        metrics['tp_gain'],
        -metrics['fp_loss'],
        -metrics['fn_opportunity_loss']
    ]

    ax1.bar(['Gains (TP)', 'Losses (FP)', 'Missed Opp. (FN)'],
            scenario_results,
            color=['green', 'red', 'orange'])
    ax1.set_title('Financial Results by Scenario')
    ax1.set_ylabel('Value ($)')
    ax1.axhline(y=0, color='black', linestyle='-', alpha=0.3)

    labels = ['TP', 'TN', 'FP', 'FN']
    sizes = [metrics['tp'], metrics['tn'], metrics['fp'], metrics['fn']]
    colors = ['green', 'blue', 'red', 'orange']

    ax2.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
    ax2.set_title('Prediction Distribution')

    plt.tight_layout()
    plt.show()

def optimize_threshold(y_true, y_proba, average_loan_amount=10000):
    """
    Advanced threshold optimization considering multiple criteria.
    
    Args:
        y_true: Actual labels
        y_proba: Predicted probabilities
        average_loan_amount: Average loan amount
        
    Returns:
        float: Optimal threshold value
    """
    thresholds = np.arange(0.3, 0.8, 0.02)
    results = []
    
    for thresh in thresholds:
        y_pred = (y_proba >= thresh).astype(int)
        
        # Calculate financial metrics
        metrics = calculate_financial_metrics(y_true, y_pred, average_loan_amount)
        
        # Calculate classification metrics
        recall_0 = recall_score(y_true, y_pred, pos_label=0)
        recall_1 = recall_score(y_true, y_pred, pos_label=1)
        
        # Composite score
        composite_score = (
            metrics['roi'] * 0.5 +  # Higher weight for ROI
            recall_0 * 0.3 +        # Significant weight for class 0 recall
            recall_1 * 0.2          # Lower weight for class 1 recall
        )
        
        results.append({
            'threshold': thresh,
            'roi': metrics['roi'],
            'recall_0': recall_0,
            'recall_1': recall_1,
            'composite_score': composite_score,
            'net_result': metrics['net_result']
        })
    
    # Convert to DataFrame for analysis
    results_df = pd.DataFrame(results)
    
    # Find best threshold
    best_threshold = results_df.loc[results_df['composite_score'].idxmax(), 'threshold']
    
    # Plot results
    plt.figure(figsize=(12, 6))
    
    # ROI plot
    plt.subplot(1, 2, 1)
    plt.plot(results_df['threshold'], results_df['roi'], 'b-', label='ROI')
    plt.axvline(x=best_threshold, color='r', linestyle='--', label='Optimal Threshold')
    plt.xlabel('Threshold')
    plt.ylabel('ROI (%)')
    plt.title('ROI vs Threshold')
    plt.legend()
    
    # Recalls plot
    plt.subplot(1, 2, 2)
    plt.plot(results_df['threshold'], results_df['recall_0'], 'g-', label='Class 0 Recall')
    plt.plot(results_df['threshold'], results_df['recall_1'], 'b-', label='Class 1 Recall')
    plt.axvline(x=best_threshold, color='r', linestyle='--', label='Optimal Threshold')
    plt.xlabel('Threshold')
    plt.ylabel('Recall')
    plt.title('Recalls vs Threshold')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Print detailed results
    best_result = results_df[results_df['threshold'] == best_threshold].iloc[0]
    print("\n=== Threshold Optimization Results ===")
    print(f"Best threshold: {best_threshold:.3f}")
    print(f"Expected ROI: {best_result['roi']:.2f}%")
    print(f"Class 0 Recall: {best_result['recall_0']:.3f}")
    print(f"Class 1 Recall: {best_result['recall_1']:.3f}")
    
    return best_threshold

def adjust_dynamic_class_weights(y, financial_metrics):
    """Adjust class weights based on financial costs"""
    fn_cost = financial_metrics['fn_opportunity_loss'] / max(financial_metrics['fn'], 1)
    fp_cost = financial_metrics['fp_loss'] / max(financial_metrics['fp'], 1)
    
    current_roi = financial_metrics['roi']
    target_roi = 8.0
    roi_factor = min(2.0, max(1.0, target_roi / max(current_roi, 1.0)))
    
    total_cost = fp_cost * 1.2 + fn_cost
    class_0_weight = (fp_cost * 1.2) / total_cost
    class_1_weight = fn_cost / total_cost
    
    counts = np.bincount(y)
    n_samples = len(y)
    
    balance_0 = n_samples / (len(np.unique(y)) * counts[0])
    balance_1 = n_samples / (len(np.unique(y)) * counts[1])
    
    balance_0 *= roi_factor
    
    weights = {
        0: class_0_weight * balance_0,
        1: class_1_weight * balance_1
    }
    
    max_weight = max(weights.values())
    weights = {k: v/max_weight for k, v in weights.items()}
    
    if weights[0] < weights[1]:
        print("\n[WARNING] Minority class weight lower than majority!")
        print("[WARNING] Adjusting minority class weight...")
        weights[0] = max(weights[1] * 1.2, weights[0])
    
    visualize_weights(weights, class_0_weight, class_1_weight, balance_0, balance_1)
    
    return {
        'weights': weights,
        'components': {
            'costs': {'0': class_0_weight, '1': class_1_weight},
            'balance': {'0': balance_0, '1': balance_1},
            'roi_factor': roi_factor
        }
    }

def visualize_weights(weights, class_0_weight, class_1_weight, balance_0, balance_1):
    """Visualize weight distributions"""
    print("\n=== Calculated Dynamic Weights ===")
    print(f"Class 0 (Not Paid): {weights[0]:.4f}")
    print(f"Class 1 (Paid): {weights[1]:.4f}")
    
    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 2, 1)
    plt.bar(['Class 0', 'Class 1'], [weights[0], weights[1]], 
            color=['red', 'blue'])
    plt.title('Weight Distribution')
    plt.ylabel('Weight')
    
    plt.subplot(1, 2, 2)
    components = pd.DataFrame({
        'Cost': [class_0_weight, class_1_weight],
        'Balance': [balance_0, balance_1]
    }, index=['Class 0', 'Class 1'])
    components.plot(kind='bar', ax=plt.gca())
    plt.title('Weight Components')
    plt.ylabel('Value')
    
    plt.tight_layout()
    plt.show()

def analyze_imbalance_impact(X_train, y_train, X_val, y_val):
    """
    Analyzes imbalance impact considering business metrics and dynamic weights
    """
    results = {}
    class_distribution = pd.Series(y_train).value_counts(normalize=True)
    print(f"\n[INFO] Original class distribution:\n{class_distribution}")

    try:
        # Base model training
        base_model = AdaBoostClassifier(
            estimator=DecisionTreeClassifier(max_depth=3),
            n_estimators=100,
            learning_rate=0.1,
            random_state=42
        )
        base_model.fit(X_train, y_train)
        model_orig = base_model

        # Initial metrics calculation
        y_pred_initial = model_orig.predict(X_val)
        initial_financial_metrics = calculate_financial_metrics(y_val, y_pred_initial)

        # SMOTE approach
        print("[INFO] Testing SMOTE...")
        smote = SMOTE(random_state=42)
        X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
        model_smote = AdaBoostClassifier(
            estimator=DecisionTreeClassifier(max_depth=3),
            n_estimators=100,
            learning_rate=0.1,
            random_state=42
        )
        model_smote.fit(X_train_smote, y_train_smote)

        # Traditional weights approach
        print("[INFO] Testing traditional weights...")
        weights_traditional = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
        sample_weights_trad = np.where(y_train == 1, weights_traditional[1], weights_traditional[0])
        model_weight_trad = AdaBoostClassifier(
            estimator=DecisionTreeClassifier(max_depth=3),
            n_estimators=100,
            learning_rate=0.1,
            random_state=42
        )
        model_weight_trad.fit(X_train, y_train, sample_weight=sample_weights_trad)

        # Dynamic weights approach
        print("[INFO] Testing dynamic weights...")
        weight_results = adjust_dynamic_class_weights(y_train, initial_financial_metrics)
        dynamic_weights = weight_results['weights']
        sample_weights_dynamic = np.array([dynamic_weights[y] for y in y_train])

        model_weight_dynamic = AdaBoostClassifier(
            estimator=DecisionTreeClassifier(max_depth=3),
            n_estimators=100,
            learning_rate=0.1,
            random_state=42
        )
        model_weight_dynamic.fit(X_train, y_train, sample_weight=sample_weights_dynamic)

        # Evaluate approaches
        approaches = {
            'Original': model_orig,
            'SMOTE': model_smote,
            'Traditional Weights': model_weight_trad,
            'Dynamic Weights': model_weight_dynamic
        }

        comparison = {}
        roi_values = {}

        for name, model in approaches.items():
            print(f"\n[INFO] Evaluating approach: {name}")
            y_pred = model.predict(X_val)
            y_proba = model.predict_proba(X_val)[:, 1]

            metrics = {
                'F1': f1_score(y_val, y_pred),
                'AUC': roc_auc_score(y_val, y_proba)
            }

            financials = calculate_financial_metrics(y_val, y_pred)
            metrics.update({
                'ROI': financials['roi'],
                'Net_Result': financials['net_result'],
                'FP_Cost': financials['fp_loss'],
                'FN_Cost': financials['fn_opportunity_loss']
            })

            comparison[name] = metrics
            roi_values[name] = metrics['ROI']

        results_df = pd.DataFrame(comparison).round(4)
        best_approach = max(roi_values.items(), key=lambda x: x[1])[0]

        # Visualize results
        try:
            plt.figure(figsize=(12, 6))
            metrics_to_plot = ['F1', 'ROI']

            for i, metric in enumerate(metrics_to_plot):
                plt.subplot(1, 2, i+1)
                values = [comparison[approach][metric] for approach in approaches.keys()]
                plt.bar(approaches.keys(), values)
                plt.title(f'{metric} Comparison')
                plt.xticks(rotation=45)
                if metric == 'ROI':
                    plt.ylabel('ROI (%)')

            plt.tight_layout()
            plt.show()
        except Exception as e:
            print(f"[WARNING] Error generating plots: {str(e)}")

        print("\n=== Imbalance Analysis ===")
        print("\nApproach comparison:")
        print(results_df)
        print(f"\nBest approach based on ROI: {best_approach}")
        print(f"Best approach ROI: {roi_values[best_approach]:.2f}%")

        return {
            'comparison': results_df,
            'best_approach': best_approach,
            'class_distribution': class_distribution,
            'best_weights': dynamic_weights if best_approach == 'Dynamic Weights' else None,
            'roi_values': roi_values,
            'models': approaches
        }

    except Exception as e:
        print(f"[ERROR] An error occurred during imbalance analysis: {str(e)}")
        print("[INFO] Returning default approach (Original)")
        return {
            'best_approach': 'Original',
            'class_distribution': class_distribution,
            'best_weights': None,
            'error': str(e)
        }

def select_important_features(X_train, y_train, threshold=0.01):
    """
    Selects most important features using a simple model
    
    Args:
        X_train: training features
        y_train: training labels
        threshold: minimum importance threshold
        
    Returns:
        list: selected important features
    """
    print("[INFO] Selecting important features...")

    # Train quick model for importance evaluation
    quick_estimator = AdaBoostClassifier(
        estimator=DecisionTreeClassifier(max_depth=2),
        n_estimators=100,
        learning_rate=0.1,
        random_state=42
    )

    quick_estimator.fit(X_train, y_train)

    # Calculate feature importance
    importance = pd.DataFrame({
        'feature': X_train.columns,
        'importance': quick_estimator.feature_importances_
    })

    # Select important features
    important_features = importance[importance['importance'] > threshold]['feature'].tolist()

    print(f"[INFO] Selected {len(important_features)} features from {X_train.shape[1]}")
    return important_features

def evaluate_unified_model(y_true, y_pred, y_proba=None, verbose=True):
    """
    Unified evaluation function combining all metrics and visualizations
    
    Args:
        y_true: true labels
        y_pred: predicted labels
        y_proba: prediction probabilities
        verbose: whether to print detailed output
        
    Returns:
        dict: evaluation results
    """
    results = {}

    # 1. Basic Classification Metrics
    if verbose:
        print("\n=== Detailed Model Evaluation ===")
        print("\nClassification Report:")
        print(classification_report(y_true, y_pred))

    # 2. Per-Class Metrics
    class_metrics = calculate_metrics_by_class(y_true, y_pred, y_proba)
    results['class_metrics'] = class_metrics

    if verbose:
        for class_name, metrics in class_metrics.items():
            print(f"\nMetrics for {class_name}:")
            for metric, value in metrics.items():
                if metric in ['TP', 'TN', 'FP', 'FN']:
                    print(f"{metric}: {value:,d}")
                else:
                    print(f"{metric}: {value:.4f}")

    # 3. Confusion Matrix
    if verbose:
        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title('Confusion Matrix with Absolute Values')
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.show()

        # Normalized matrix
        cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm_norm, annot=True, fmt='.2%', cmap='Blues')
        plt.title('Normalized Confusion Matrix')
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.show()

    # 4. Financial Metrics
    financial_metrics = calculate_financial_metrics(y_true, y_pred)
    results['financial_metrics'] = financial_metrics

    # 5. Comparative Visualizations
    if verbose:
        create_comparative_plots(class_metrics, y_true, y_pred, y_proba)

    return results

def run_optimization_pipeline(train_path, test_path, sample_size=None):
    """
    Executes initial pipeline with optimized sample to find best hyperparameters.
    Uses a hybrid approach: optimizes parameters except n_estimators.

    Args:
        train_path (str): Path to training data file
        test_path (str): Path to test data file 
        sample_size (int, optional): Size of sample to use. If None, calculated automatically.

    Returns:
        dict: Dictionary containing optimization results including:
            - best_params: Best hyperparameters found
            - financial_metrics: Financial performance metrics
            - na_analysis: Missing value analysis results
            - balancing_analysis: Class balancing analysis results
            - best_approach: Best balancing approach identified
    """
    print("\n[INFO] Starting optimization pipeline...")

    # 1. Load and Initial Exploration
    df_train, df_test = load_data(train_path, test_path)

    # Stratified sampling
    if sample_size is None:
        class_counts = df_train['y'].value_counts()
        min_class_count = class_counts.min()
        sample_size = max(min_class_count * 3, int(len(df_train) * 0.3))
        print(f"[INFO] Calculated sample size: {sample_size:,.0f} records")

    df_train_sample = df_train.groupby('y', group_keys=False).apply(
        lambda x: x.sample(n=min(len(x), int(sample_size * len(x)/len(df_train))))
    )

    complete_data_exploration(df_train_sample, "Training Sample")
    analyze_target_distribution(df_train_sample)

    # 2. Detailed missing value analysis
    print("\n[INFO] Performing detailed missing value analysis...")
    na_analysis = analyze_missing_values_detailed(df_train_sample, df_test)

    # Apply NA analysis recommendations
    features_keep = df_train_sample.columns.difference(na_analysis['features_to_drop'])
    df_train_sample = df_train_sample[features_keep]
    df_test = df_test[features_keep]

    # 3. Correlation Analysis and variable removal
    variables_remove = analyze_correlations(df_train_sample)
    columns_keep = [col for col in df_train_sample.columns if col not in variables_remove]
    df_train_sample = df_train_sample[columns_keep]

    # 4. Data Treatment
    df_train_sample = handle_outliers(df_train_sample)
    df_train_sample, _ = handle_missing_values(df_train_sample, df_train_sample.copy())

    # 5. Modeling Preparation
    X_sample = df_train_sample.drop('y', axis=1)
    y_sample = df_train_sample['y']
    X_train, X_val, y_train, y_val = train_test_split(
        X_sample,
        y_sample,
        test_size=0.2,
        random_state=42,
        stratify=y_sample
    )

    # 6. Imbalance Analysis
    print("\n[INFO] Analyzing imbalance impact...")
    balancing_analysis = analyze_imbalance_impact(
        X_train, y_train, X_val, y_val
    )

    # Apply recommended balancing strategy
    if balancing_analysis['best_approach'] == 'SMOTE':
        print("[INFO] Applying SMOTE for balancing...")
        X_train, y_train = SMOTE(random_state=42).fit_resample(X_train, y_train)
        sample_weights = None
    elif balancing_analysis['best_approach'] == 'Class Weights':
        print("[INFO] Using traditional class weights...")
        weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
        sample_weights = np.where(y_train == 1, weights[1], weights[0])
    elif balancing_analysis['best_approach'] == 'Dynamic Weights':
        print("[INFO] Applying dynamic weights...")
        dynamic_weights = balancing_analysis['best_weights']
        if dynamic_weights:
            sample_weights = np.array([dynamic_weights[y] for y in y_train])
        else:
            print("[WARNING] Dynamic weights unavailable, using traditional balancing...")
            weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
            sample_weights = np.where(y_train == 1, weights[1], weights[0])
    else:
        print("[INFO] Keeping original data without balancing...")
        sample_weights = None

    # 7. Selection important features
    important_features = select_important_features(X_train, y_train)
    X_train = X_train[important_features]
    X_val = X_val[important_features]

    # 8. Hyperparameter Optimization - Reduced Grid
    param_grid = {
        'n_estimators': [200, 250, 300],
        'learning_rate': [0.1, 0.3],
        'estimator__max_depth': [3, 4],
        'estimator__min_samples_leaf': [50, 100],
        'algorithm': ['SAMME.R']
    }

    # Base model configuration
    base_estimator = DecisionTreeClassifier()
    model = AdaBoostClassifier(
        estimator=base_estimator,
        random_state=42
    )

    # Stratified cross-validation
    cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

    # RandomizedSearchCV
    random_search = RandomizedSearchCV(
        model,
        param_distributions=param_grid,
        n_iter=10,
        cv=cv,
        scoring='f1',
        n_jobs=-1,
        verbose=2,
        random_state=42
    )

    print("[INFO] Starting Random Search on sample...")
    if sample_weights is not None:
        random_search.fit(X_train, y_train, sample_weight=sample_weights)
    else:
        random_search.fit(X_train, y_train)

    # 9. Optimized Model Evaluation
    print("\n[INFO] Best parameters found:")
    print(random_search.best_params_)
    print("\n[INFO] Best score (F1):")
    print(f"{random_search.best_score_:.4f}")

    y_pred = random_search.predict(X_val)
    y_proba = random_search.predict_proba(X_val)[:, 1]

    print("\n[INFO] Optimized model evaluation on sample:")
    results = evaluate_unified_model(
        y_val,
        y_pred,
        y_proba,
        verbose=True
    )

    # 10. Financial Analysis of Sample
    print("\n[INFO] Sample financial analysis:")
    financial_metrics = results['financial_metrics']

    # Return complete results
    return {
        'best_params': random_search.best_params_,
        'financial_metrics': financial_metrics,
        'na_analysis': na_analysis,
        'balancing_analysis': balancing_analysis,
        'important_features': important_features,
        'best_approach': balancing_analysis['best_approach']
    }

def run_complete_pipeline(train_path, test_path, optimization_results):
    """
    Executes complete pipeline with all data using optimization results.
    
    Args:
        train_path (str): Path to training data file
        test_path (str): Path to test data file
        optimization_results (dict): Results from optimization pipeline

    Returns:
        dict: Dictionary containing final model results including:
            - model: Trained final model
            - financial_metrics: Financial performance metrics
            - optimal_threshold: Optimized classification threshold
            - training_time: Model training duration
            - balancing_strategy: Strategy used for class balancing
            - na_analysis: Missing value analysis results
    """
    print("\n[INFO] Starting complete pipeline with all data...")

    # 1. Load and Initial Exploration
    df_train, df_test = load_data(train_path, test_path)
    complete_data_exploration(df_train, "Complete Training Set")
    complete_data_exploration(df_test, "Test Set")
    analyze_target_distribution(df_train)

    # 2. Detailed missing value analysis
    print("[INFO] Performing detailed missing value analysis...")
    na_analysis = analyze_missing_values_detailed(df_train, df_test)

    # Apply NA analysis recommendations
    features_keep = df_train.columns.difference(na_analysis['features_to_drop'])
    df_train = df_train[features_keep]
    df_test = df_test[features_keep]

    # 3. Correlation Analysis and variable removal
    variables_remove = analyze_correlations(df_train)
    columns_keep = [col for col in df_train.columns if col not in variables_remove]

    # Ensure both sets have same columns
    df_train = df_train[columns_keep]
    df_test = df_test[columns_keep]

    # 4. Data Treatment
    df_train = handle_outliers(df_train)
    df_test = handle_outliers(df_test)
    df_train, df_test = handle_missing_values(df_train, df_test)

    # 5. Data Preparation - Including validation set split
    X = df_train.drop('y', axis=1)
    y = df_train['y']

    # Split data into train and validation
    X_train, X_val, y_train, y_val = train_test_split(
        X, y, test_size=0.2, stratify=y, random_state=42
    )

    # Prepare test set
    X_test = df_test.drop('y', axis=1)
    y_test = df_test['y']

    # 6. Apply best balancing strategy from optimization
    print("[INFO] Applying balancing strategy based on optimization...")
    best_approach = optimization_results['best_approach']

    if best_approach == 'SMOTE':
        print("[INFO] Applying SMOTE for balancing...")
        X_train, y_train = SMOTE(random_state=42).fit_resample(X_train, y_train)
        sample_weights = None
    elif best_approach == 'Class Weights':
        print("[INFO] Using traditional class weights...")
        weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
        sample_weights = np.where(y_train == 1, weights[1], weights[0])
    elif best_approach == 'Dynamic Weights':
        print("[INFO] Applying dynamic weights from optimized model...")
        # First train initial model to get financial metrics
        initial_model = AdaBoostClassifier(
            estimator=DecisionTreeClassifier(max_depth=3),
            n_estimators=100,
            learning_rate=0.1,
            random_state=42
        )
        initial_model.fit(X_train, y_train)
        y_pred_initial = initial_model.predict(X_val)
        financial_metrics_initial = calculate_financial_metrics(y_val, y_pred_initial)

        # Calculate dynamic weights
        dynamic_weights = adjust_dynamic_class_weights(y_train, financial_metrics_initial)
        sample_weights = np.array([dynamic_weights[y] for y in y_train])
    else:
        print("[INFO] Keeping original data without balancing...")
        sample_weights = None

    # 7. Selection important features based on optimization
    important_features = optimization_results['important_features']
    X_train = X_train[important_features]
    X_val = X_val[important_features]
    X_test = X_test[important_features]

    # 8. Train Final Model
    print("[INFO] Training final model...")
    best_params = optimization_results['best_params']

    base_estimator = DecisionTreeClassifier(
        max_depth=best_params['estimator__max_depth'],
        min_samples_leaf=best_params.get('estimator__min_samples_leaf', 50)
    )

    start_time = time.time()
    final_model = AdaBoostClassifier(
        estimator=base_estimator,
        learning_rate=best_params['learning_rate'],
        n_estimators=best_params['n_estimators'],
        algorithm=best_params.get('algorithm', 'SAMME.R'),
        random_state=42
    )

    # Train model with appropriate weights
    if sample_weights is not None:
        final_model.fit(X_train, y_train, sample_weight=sample_weights)
    else:
        final_model.fit(X_train, y_train)

    training_time = time.time() - start_time
    print(f"[INFO] Training time: {training_time:.2f} seconds")

    # 9. Final Model Evaluation
    y_pred = final_model.predict(X_test)
    y_proba = final_model.predict_proba(X_test)[:, 1]

    # 11. Threshold Optimization
    print("[INFO] Optimizing classification threshold...")
    optimal_threshold = optimize_threshold(y_test, y_proba)
    y_pred_optimized = (y_proba >= optimal_threshold).astype(int)

    # 12. Complete Model Evaluation
    print("\n[INFO] Final model evaluation:")
    results = evaluate_unified_model(
        y_test,
        y_pred_optimized,
        y_proba,
        verbose=True
    )
    financial_metrics = results['financial_metrics']

    # 13. Cross Validation
    print("\n[INFO] Performing cross validation...")
    cv_scores = cross_val_score(final_model, X_test, y_test, cv=3, scoring='f1', n_jobs=-1)
    print(f"[INFO] Cross Validation Results (F1-Score): {cv_scores}")
    print(f"[INFO] Mean F1-Score in Cross Validation: {cv_scores.mean():.4f}")

    # 14. Learning Curve
    plot_learning_curve(final_model, X_test, y_test, "Learning Curve - AdaBoost")

    # 15. Generate Report
    generate_report(final_model, X_test, y_test, financial_metrics, training_time)

    # 16. Return complete results
    return {
        'model': final_model,
        'financial_metrics': financial_metrics,
        'optimal_threshold': optimal_threshold,
        'training_time': training_time,
        'features_used': important_features,
        'balancing_strategy': best_approach,
        'na_analysis': na_analysis
    }

def calculate_metrics_by_class(y_true, y_pred, y_proba=None):
    """
    Calculates detailed metrics for each class.
    
    Args:
        y_true (array-like): True labels
        y_pred (array-like): Predicted labels
        y_proba (array-like, optional): Prediction probabilities for positive class

    Returns:
        dict: Dictionary containing metrics for each class including:
            - Precision, Recall, F1-Score
            - Specificity
            - True/False Positives/Negatives
            - AUC-ROC (if probabilities provided)
    """
    # General metrics by class
    metrics = {
        'Class 0 (Not Paid)': {
            'Precision': precision_score(y_true, y_pred, pos_label=0),
            'Recall': recall_score(y_true, y_pred, pos_label=0),
            'F1-Score': f1_score(y_true, y_pred, pos_label=0),
            'Specificity': recall_score(y_true, y_pred, pos_label=1),
            'TP': confusion_matrix(y_true, y_pred)[0,0],
            'TN': confusion_matrix(y_true, y_pred)[1,1],
            'FP': confusion_matrix(y_true, y_pred)[1,0],
            'FN': confusion_matrix(y_true, y_pred)[0,1]
        },
        'Class 1 (Paid)': {
            'Precision': precision_score(y_true, y_pred, pos_label=1),
            'Recall': recall_score(y_true, y_pred, pos_label=1),
            'F1-Score': f1_score(y_true, y_pred, pos_label=1),
            'Specificity': recall_score(y_true, y_pred, pos_label=0),
            'TP': confusion_matrix(y_true, y_pred)[1,1],
            'TN': confusion_matrix(y_true, y_pred)[0,0],
            'FP': confusion_matrix(y_true, y_pred)[0,1],
            'FN': confusion_matrix(y_true, y_pred)[1,0]
        }
    }

    if y_proba is not None:
        metrics['Class 0 (Not Paid)']['AUC-ROC'] = roc_auc_score(y_true == 0, 1 - y_proba)
        metrics['Class 1 (Paid)']['AUC-ROC'] = roc_auc_score(y_true == 1, y_proba)

    return metrics

def create_comparative_plots(metrics, y_true, y_pred, y_proba=None):
    """
    Creates comparative visualizations of metrics between classes.
    
    Args:
        metrics (dict): Dictionary containing metrics for each class
        y_true (array-like): True labels
        y_pred (array-like): Predicted labels 
        y_proba (array-like, optional): Prediction probabilities for positive class
        
    Creates multiple plots:
    - Grouped bar chart comparing main metrics
    - Normalized confusion matrix heatmap
    - ROC curves for both classes
    - Distribution of TP, TN, FP, FN
    - Radar plot comparing key metrics
    """
    # Style configuration
    plt.style.use('default')
    colors = ['#ff6b6b', '#4ecdc4']

    # 1. Comparative bar plot of main metrics
    main_metrics = ['Precision', 'Recall', 'F1-Score']

    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    fig.suptitle('Metrics Comparison by Class', fontsize=16, y=1.02)

    # 1.1 Grouped bar chart
    ax = axes[0,0]
    x = np.arange(len(main_metrics))
    width = 0.35

    class0_values = [metrics['Class 0 (Not Paid)'][m] for m in main_metrics]
    class1_values = [metrics['Class 1 (Paid)'][m] for m in main_metrics]

    ax.bar(x - width/2, class0_values, width, label='Not Paid (0)', color=colors[0])
    ax.bar(x + width/2, class1_values, width, label='Paid (1)', color=colors[1])

    ax.set_ylabel('Value')
    ax.set_title('Main Metrics')
    ax.set_xticks(x)
    ax.set_xticklabels(main_metrics)
    ax.legend()

    # Add values above bars
    for i in x:
        ax.text(i - width/2, class0_values[i], f'{class0_values[i]:.3f}',
                ha='center', va='bottom')
        ax.text(i + width/2, class1_values[i], f'{class1_values[i]:.3f}',
                ha='center', va='bottom')

    # 1.2 Normalized Confusion Matrix
    ax = axes[0,1]
    cm = confusion_matrix(y_true, y_pred, normalize='true')
    sns.heatmap(cm, annot=True, fmt='.2%', cmap='Blues', ax=ax,
                xticklabels=['Not Paid (0)', 'Paid (1)'],
                yticklabels=['Not Paid (0)', 'Paid (1)'])
    ax.set_title('Normalized Confusion Matrix')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

    # 1.3 ROC Curves
    ax = axes[1,0]
    if y_proba is not None:
        # ROC for class 0
        fpr_0, tpr_0, _ = roc_curve(y_true == 0, 1 - y_proba)
        auc_0 = roc_auc_score(y_true == 0, 1 - y_proba)
        ax.plot(fpr_0, tpr_0, color=colors[0],
                label=f'Class 0 (AUC = {auc_0:.3f})')

        # ROC for class 1
        fpr_1, tpr_1, _ = roc_curve(y_true == 1, y_proba)
        auc_1 = roc_auc_score(y_true == 1, y_proba)
        ax.plot(fpr_1, tpr_1, color=colors[1],
                label=f'Class 1 (AUC = {auc_1:.3f})')

        ax.plot([0, 1], [0, 1], 'k--')
        ax.set_xlabel('False Positive Rate')
        ax.set_ylabel('True Positive Rate')
        ax.set_title('ROC Curves by Class')
        ax.legend()

    # 1.4 TP, TN, FP, FN Distribution
    ax = axes[1,1]
    confusion_metrics = ['TP', 'TN', 'FP', 'FN']
    confusion_values_0 = [metrics['Class 0 (Not Paid)'][m] for m in confusion_metrics]
    confusion_values_1 = [metrics['Class 1 (Paid)'][m] for m in confusion_metrics]

    x_conf = np.arange(len(confusion_metrics))
    ax.bar(x_conf - width/2, confusion_values_0, width, label='Not Paid (0)', color=colors[0])
    ax.bar(x_conf + width/2, confusion_values_1, width, label='Paid (1)', color=colors[1])

    ax.set_ylabel('Count')
    ax.set_title('Results Distribution')
    ax.set_xticks(x_conf)
    ax.set_xticklabels(confusion_metrics)
    ax.legend()

    plt.tight_layout()
    plt.show()

    # 2. Radar Plot (Spider Plot)
    plt.figure(figsize=(10, 10))

    # Prepare data for radar plot
    radar_metrics = ['Precision', 'Recall', 'F1-Score']
    radar_values_0 = [metrics['Class 0 (Not Paid)'][m] for m in radar_metrics]
    radar_values_1 = [metrics['Class 1 (Paid)'][m] for m in radar_metrics]

    # Configure radar plot
    angles = np.linspace(0, 2*np.pi, len(radar_metrics), endpoint=False)
    radar_values_0 = np.concatenate((radar_values_0, [radar_values_0[0]]))
    radar_values_1 = np.concatenate((radar_values_1, [radar_values_1[0]]))
    angles = np.concatenate((angles, [angles[0]]))

    ax = plt.subplot(111, projection='polar')
    ax.plot(angles, radar_values_0, 'o-', linewidth=2, label='Not Paid (0)', color=colors[0])
    ax.fill(angles, radar_values_0, alpha=0.25, color=colors[0])
    ax.plot(angles, radar_values_1, 'o-', linewidth=2, label='Paid (1)', color=colors[1])
    ax.fill(angles, radar_values_1, alpha=0.25, color=colors[1])

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(radar_metrics)
    ax.set_title('Radial Metrics Comparison')
    ax.grid(True)
    ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

    plt.show()

def main():
    """
    Main execution function that runs the complete pipeline:
    1. Calculates sample size
    2. Runs optimization phase with reduced sample
    3. Runs complete training with optimized settings
    4. Outputs final summary of results
    
    Returns:
        dict: Final results containing training time, ROI, balancing strategy and features used
    """
    train_path = 'train.csv'
    test_path = 'test.csv'

    # Calculate sample size as 30% of data
    print("[INFO] Calculating sample size...")
    df_temp = pd.read_csv(train_path)
    sample_size = int(len(df_temp) * 0.3)
    del df_temp  # Free memory

    # Phase 1: Optimization with reduced sample
    print("\n[INFO] Starting optimization phase...")
    optimization_results = run_optimization_pipeline(
        train_path,
        test_path,
        sample_size=sample_size
    )

    # Phase 2: Complete training with optimized settings
    print("\n[INFO] Starting complete training with optimized settings...")
    final_results = run_complete_pipeline(
        train_path,
        test_path,
        optimization_results
    )

    print("\n[INFO] Complete pipeline finished successfully!")

    # Final summary
    print("\n=== Final Summary ===")
    print(f"Total training time: {final_results['training_time']:.2f} seconds")
    print(f"Final ROI: {final_results['financial_metrics']['roi']:.2f}%")
    print(f"Balancing strategy used: {final_results['balancing_strategy']}")
    print(f"Number of features used: {len(final_results['features_used'])}")

    return final_results
```

```python
def plot_financial_comparison(metrics_ada, metrics_cat):
    """
    Creates detailed financial metrics comparison visualizations
    """
    fig = plt.figure(figsize=(20, 15))
    gs = GridSpec(2, 3, figure=fig)

    # 1. ROI Comparison
    ax1 = fig.add_subplot(gs[0, 0])
    roi_data = [metrics_ada['roi'], metrics_cat['roi']]
    colors = ['skyblue', 'lightgreen']
    bars = ax1.bar(['AdaBoost', 'CatBoost'], roi_data, color=colors)
    ax1.set_title('Return on Investment (ROI)', fontsize=12, pad=15)
    ax1.set_ylabel('ROI (%)')
    
    # Add value labels
    for bar in bars:
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.2f}%', ha='center', va='bottom')

    # 2. Net Result Comparison
    ax2 = fig.add_subplot(gs[0, 1])
    net_results = [metrics_ada['net_result'], metrics_cat['net_result']]
    bars = ax2.bar(['AdaBoost', 'CatBoost'], net_results, color=colors)
    ax2.set_title('Net Financial Result', fontsize=12, pad=15)
    ax2.set_ylabel('Amount ($)')
    
    # Add value labels
    for bar in bars:
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height,
                f'${height:,.0f}', ha='center', va='bottom')

    # 3. Gains vs Losses
    ax3 = fig.add_subplot(gs[0, 2])
    models = ['AdaBoost', 'CatBoost']
    gains = [metrics_ada['total_gain'], metrics_cat['total_gain']]
    losses = [metrics_ada['total_loss'], metrics_cat['total_loss']]
    
    x = np.arange(len(models))
    width = 0.35
    
    ax3.bar(x - width/2, gains, width, label='Total Gains', color='green', alpha=0.6)
    ax3.bar(x + width/2, losses, width, label='Total Losses', color='red', alpha=0.6)
    
    ax3.set_xticks(x)
    ax3.set_xticklabels(models)
    ax3.set_title('Gains vs Losses Comparison', fontsize=12, pad=15)
    ax3.set_ylabel('Amount ($)')
    ax3.legend()

    # Add value labels
    for i, v in enumerate(gains):
        ax3.text(i - width/2, v, f'${v:,.0f}', ha='center', va='bottom')
    for i, v in enumerate(losses):
        ax3.text(i + width/2, v, f'${v:,.0f}', ha='center', va='bottom')

    # 4. Prediction Distribution with Financial Impact
    ax4 = fig.add_subplot(gs[1, :])
    
    categories = ['True Positives', 'True Negatives', 'False Positives', 'False Negatives']
    ada_values = [metrics_ada['tp'], metrics_ada['tn'], metrics_ada['fp'], metrics_ada['fn']]
    cat_values = [metrics_cat['tp'], metrics_cat['tn'], metrics_cat['fp'], metrics_cat['fn']]
    
    x = np.arange(len(categories))
    width = 0.35
    
    # Calculate financial impact per category
    loan_amount = 10000
    return_rate = 0.20
    loss_rate = 0.70
    
    def calculate_impact(count, category):
        if category == 'True Positives':
            return count * loan_amount * return_rate
        elif category == 'False Positives':
            return -count * loan_amount * loss_rate
        elif category == 'False Negatives':
            return -count * loan_amount * return_rate
        else:  # True Negatives
            return 0
    
    ada_impacts = [calculate_impact(v, c) for v, c in zip(ada_values, categories)]
    cat_impacts = [calculate_impact(v, c) for v, c in zip(cat_values, categories)]
    
    # Create bars
    bars1 = ax4.bar(x - width/2, ada_values, width, label='AdaBoost')
    bars2 = ax4.bar(x + width/2, cat_values, width, label='CatBoost')
    
    # Customize the plot
    ax4.set_ylabel('Number of Predictions')
    ax4.set_title('Prediction Distribution with Financial Impact', fontsize=12, pad=15)
    ax4.set_xticks(x)
    ax4.set_xticklabels(categories)
    ax4.legend()
    
    # Add count labels
    def autolabel(bars, impacts):
        for idx, rect in enumerate(bars):
            height = rect.get_height()
            impact = impacts[idx]
            if impact != 0:
                impact_text = f'(${impact:,.0f})' if impact < 0 else f'(+${impact:,.0f})'
            else:
                impact_text = ''
            
            ax4.text(rect.get_x() + rect.get_width()/2., height,
                    f'{int(height):,}\n{impact_text}',
                    ha='center', va='bottom')
    
    autolabel(bars1, ada_impacts)
    autolabel(bars2, cat_impacts)

    plt.tight_layout()
    plt.show()

def compare_models(y_test, y_pred_ada, y_pred_cat, y_proba_ada, y_proba_cat, time_ada, time_cat):
    """
    Compares AdaBoost and CatBoost models using classification and financial metrics
    """
    # Calculate classification metrics
    metrics = {
        'AdaBoost': {
            'Precision': precision_score(y_test, y_pred_ada),
            'Recall': recall_score(y_test, y_pred_ada), 
            'F1-Score': f1_score(y_test, y_pred_ada),
            'ROC AUC': roc_auc_score(y_test, y_proba_ada),
            'Time': time_ada
        },
        'CatBoost': {
            'Precision': precision_score(y_test, y_pred_cat),
            'Recall': recall_score(y_test, y_pred_cat),
            'F1-Score': f1_score(y_test, y_pred_cat),
            'ROC AUC': roc_auc_score(y_test, y_proba_cat),
            'Time': time_cat
        }
    }
    
    # Calculate financial metrics for both models
    fin_metrics_ada = calculate_financial_metrics(y_test, y_pred_ada)
    fin_metrics_cat = calculate_financial_metrics(y_test, y_pred_cat)
    
    # Add financial metrics to comparison
    metrics['AdaBoost'].update({
        'ROI': fin_metrics_ada['roi'],
        'Net Result': fin_metrics_ada['net_result'],
        'Total Gain': fin_metrics_ada['total_gain'],
        'Total Loss': fin_metrics_ada['total_loss']
    })
    
    metrics['CatBoost'].update({
        'ROI': fin_metrics_cat['roi'],
        'Net Result': fin_metrics_cat['net_result'],
        'Total Gain': fin_metrics_cat['total_gain'],
        'Total Loss': fin_metrics_cat['total_loss']
    })

    plot_financial_comparison(fin_metrics_ada, fin_metrics_cat)

    # Visualizations
    fig = plt.figure(figsize=(20, 15))
    gs = GridSpec(3, 3, figure=fig)
    
    # 1. Classification Metrics
    ax1 = fig.add_subplot(gs[0, :2])
    class_metrics_df = pd.DataFrame(metrics).loc[['Precision', 'Recall', 'F1-Score', 'ROC AUC']]
    class_metrics_df.plot(kind='bar', ax=ax1)
    ax1.set_title('Classification Metrics Comparison')
    ax1.set_ylabel('Value')
    plt.xticks(rotation=45)
    
    # Add values on bars
    for i in range(len(class_metrics_df.index)):
        for j in range(len(class_metrics_df.columns)):
            ax1.text(i - 0.2 + j*0.4, class_metrics_df.iloc[i, j], 
                    f'{class_metrics_df.iloc[i, j]:.3f}', 
                    ha='center', va='bottom')
    
    # 2. Financial Metrics
    ax2 = fig.add_subplot(gs[0, 2])
    fin_metrics_df = pd.DataFrame(metrics).loc[['ROI', 'Net Result']]
    fin_metrics_df.plot(kind='bar', ax=ax2)
    ax2.set_title('Financial Metrics Comparison')
    ax2.set_ylabel('Value')
    plt.xticks(rotation=45)
    
    # Add values on bars
    for i in range(len(fin_metrics_df.index)):
        for j in range(len(fin_metrics_df.columns)):
            value = fin_metrics_df.iloc[i, j]
            if fin_metrics_df.index[i] == 'ROI':
                text = f'{value:.2f}%'
            else:
                text = f'${value:,.0f}'
            ax2.text(i - 0.2 + j*0.4, value, text, ha='center', va='bottom')
    
    # 3. Confusion Matrices
    ax3 = fig.add_subplot(gs[1, 0])
    cm_ada = confusion_matrix(y_test, y_pred_ada, normalize='true')
    sns.heatmap(cm_ada, annot=True, fmt='.2%', cmap='Blues', ax=ax3,
                xticklabels=['Not Paid', 'Paid'],
                yticklabels=['Not Paid', 'Paid'])
    ax3.set_title('Confusion Matrix - AdaBoost')
    ax3.set_xlabel('Predicted')
    ax3.set_ylabel('Actual')
    
    ax4 = fig.add_subplot(gs[1, 1])
    cm_cat = confusion_matrix(y_test, y_pred_cat, normalize='true')
    sns.heatmap(cm_cat, annot=True, fmt='.2%', cmap='Blues', ax=ax4,
                xticklabels=['Not Paid', 'Paid'],
                yticklabels=['Not Paid', 'Paid'])
    ax4.set_title('Confusion Matrix - CatBoost')
    ax4.set_xlabel('Predicted')
    ax4.set_ylabel('Actual')
    
    # 4. ROC Curves
    ax5 = fig.add_subplot(gs[1, 2])
    fpr_ada, tpr_ada, _ = roc_curve(y_test, y_proba_ada)
    fpr_cat, tpr_cat, _ = roc_curve(y_test, y_proba_cat)
    
    ax5.plot(fpr_ada, tpr_ada, label=f'AdaBoost (AUC = {metrics["AdaBoost"]["ROC AUC"]:.3f})')
    ax5.plot(fpr_cat, tpr_cat, label=f'CatBoost (AUC = {metrics["CatBoost"]["ROC AUC"]:.3f})')
    ax5.plot([0, 1], [0, 1], 'k--')
    ax5.set_xlabel('False Positive Rate')
    ax5.set_ylabel('True Positive Rate')
    ax5.set_title('ROC Curves')
    ax5.legend()
    
    # 5. Financial Comparison
    ax6 = fig.add_subplot(gs[2, :])
    fin_comparison = pd.DataFrame({
        'AdaBoost': [fin_metrics_ada['tp'], fin_metrics_ada['tn'], 
                    fin_metrics_ada['fp'], fin_metrics_ada['fn']],
        'CatBoost': [fin_metrics_cat['tp'], fin_metrics_cat['tn'], 
                    fin_metrics_cat['fp'], fin_metrics_cat['fn']]
    }, index=['True Positives', 'True Negatives', 'False Positives', 'False Negatives'])
    
    fin_comparison.plot(kind='bar', ax=ax6)
    ax6.set_title('Prediction Distribution Comparison')
    ax6.set_ylabel('Count')
    
    plt.tight_layout()
    plt.show()
    
    # Results Analysis
    print("\n=== Detailed Comparative Analysis ===")
    
    print("\n1. Classification Performance:")
    class_comparison = pd.DataFrame(metrics).loc[['Precision', 'Recall', 'F1-Score', 'ROC AUC', 'Time']].round(4)
    print(class_comparison)
    
    print("\n2. Financial Performance:")
    fin_comparison = pd.DataFrame({
        'AdaBoost': {
            'ROI (%)': fin_metrics_ada['roi'],
            'Net Result ($)': fin_metrics_ada['net_result'],
            'Total Gain ($)': fin_metrics_ada['total_gain'],
            'Total Loss ($)': fin_metrics_ada['total_loss'],
            'TP Count': fin_metrics_ada['tp'],
            'TN Count': fin_metrics_ada['tn'],
            'FP Count': fin_metrics_ada['fp'],
            'FN Count': fin_metrics_ada['fn']
        },
        'CatBoost': {
            'ROI (%)': fin_metrics_cat['roi'],
            'Net Result ($)': fin_metrics_cat['net_result'],
            'Total Gain ($)': fin_metrics_cat['total_gain'],
            'Total Loss ($)': fin_metrics_cat['total_loss'],
            'TP Count': fin_metrics_cat['tp'],
            'TN Count': fin_metrics_cat['tn'],
            'FP Count': fin_metrics_cat['fp'],
            'FN Count': fin_metrics_cat['fn']
        }
    }).round(2)
    print(fin_comparison)
    
    # Model Recommendation
    print("\n3. Final Recommendations:")
    # Classification performance
    better_classifier = 'AdaBoost' if metrics['AdaBoost']['F1-Score'] > metrics['CatBoost']['F1-Score'] else 'CatBoost'
    # Financial performance
    better_financial = 'AdaBoost' if fin_metrics_ada['roi'] > fin_metrics_cat['roi'] else 'CatBoost'
    # Speed
    faster_model = 'AdaBoost' if time_ada < time_cat else 'CatBoost'
    
    print(f"\nBest Classification Performance: {better_classifier}")
    print(f"Best Financial Performance: {better_financial}")
    print(f"Faster Training: {faster_model}")
    
    print("\nKey Insights:")
    print(f"- Classification difference (F1-Score): {abs(metrics['AdaBoost']['F1-Score'] - metrics['CatBoost']['F1-Score']):.4f}")
    print(f"- ROI difference: {abs(fin_metrics_ada['roi'] - fin_metrics_cat['roi']):.2f}%")
    print(f"- Training time difference: {abs(time_ada - time_cat):.2f} seconds")
    
    return metrics, class_comparison, fin_comparison
```
